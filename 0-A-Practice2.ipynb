```python
import cv2
import matplotlib.pyplot as plt

# read 
img = cv2.imread('./images/input.jpg')

# properties
print(img.shape)

# show
# method 1: cv2
cv2.imshow('Its a document',img)
cv2.waitKey()
cv2.destroyAllWindows()
# method 2: plt
plt.imshow(img)


# write
cv2.imwrite('img2.jpg', img)


# manipulate img array
# method 1: cv2
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_rgb[200:300, 300:500, ::] = (255, 0, 0)
# method 2: slicing
img_rgb = img[..., ::-1]
img_rgb[200:300, 300:500, ::] = (0, 0, 255)


plt.imshow(img_rgb)


```
